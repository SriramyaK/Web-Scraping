```python
from selenium import webdriver
import requests
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup
import re
import pandas as pd
import os
```

```python
Program_Name = []
Details = []

url = 'https://a816-healthpsi.nyc.gov/ChildCare/SearchAction2.do'
driver = webdriver.Firefox()
driver.implicitly_wait(30)
driver.get(url)
   
u_element = driver.find_elements_by_xpath('//*[@id="ageRanges"]/option[2]') #FHSU
u_element[0].click()

u_Borough = driver.find_element_by_xpath("//select[@name='borough']/option[text()='Queens']")
u_Borough.click()

search_click = driver.find_elements_by_xpath('//*[@id="left-nav"]/section/form/div/div[2]/div/div[3]/div[3]/input[1]')
search_click[0].click()

table =  driver.find_element_by_xpath("//table[@class='enhanced']")
    
    
for row1 in table.find_elements_by_xpath(".//tr"):
    Program_Name.append([td.text for td in row1.find_elements_by_xpath(".//td[@class='persist essential']")])

for row in table.find_elements_by_xpath(".//tr"):
    Details.append([td.text for td in row.find_elements_by_xpath(".//td[@class='optional']")])
    print("First Page")

l1 = [20,30,40,50,60,70]

for i in l1:
    url1 = '?pager.offset=' + str(i)
    urlf = url + url1
    print(urlf)
    driver.get(urlf)
         
    table =  driver.find_element_by_xpath("//table[@class='enhanced']")
        
    for row1 in table.find_elements_by_xpath(".//tr"):
        Program_Name.append([td.text for td in row1.find_elements_by_xpath(".//td[@class='persist essential']")])

    for row in table.find_elements_by_xpath(".//tr"):
        Details.append([td.text for td in row.find_elements_by_xpath(".//td[@class='optional']")])
        print("Everything Fine")
        


```

```python
Final = pd.DataFrame({'Program Name': Program_Name,
                              'Details': Details},
                              columns = ['Program Name','Details']
                             )

print(Final)
```

```python
Final.to_csv('Child_care_data_NYC_Ver1.0.csv',index=False, encoding='utf-8')
```
