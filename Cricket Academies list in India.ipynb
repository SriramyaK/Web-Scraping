```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
from lxml import html
```

```python
r = requests.get('https://www.worldblaze.in/best-cricket-academies-india/')
soup = BeautifulSoup(r.text,'html.parser')
Main = soup.find_all('h3')

CA=[]

for name in Main[:-3]:
    CA.append(name.text)
    
df = pd.DataFrame(CA)
df
```
