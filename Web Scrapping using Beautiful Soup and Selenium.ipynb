```python
from bs4 import BeautifulSoup
from selenium import webdriver
import requests
import pandas as pd
import time
```

```python
url = 'https://www.indiacom.com/yellow-pages/departmental-stores/indore/'

html = requests.get(url)
IM_Soup = BeautifulSoup(html.text,'html.parser')

rooturl = "https://www.indiacom.com/"
PhNo = []
Title = []
for link in IM_Soup.findAll('a'):
        
    if link != None and link.get('href') != None:
        str1 = str(link.get("href"))
        if str1.find("/indore/") != -1:
            browser = webdriver.Firefox()
            url1 = rooturl + link.get('href')
            #print("url1: " + url1)
            browser.get(url1)
            
            elem = browser.find_element_by_xpath('//*[@id="div_details"]/div[1]/h1')
            Title.append(elem.text)
            print(Title)
            elem = browser.find_element_by_xpath('//*[@id="btn_phone"]/a')
            
            elem.click()
            time.sleep(0.5)
            
            elem = browser.find_element_by_xpath('//*[@id="div_phoneadd"]/div[1]/strong/a') 
            PhNo.append(elem.text)
            print(PhNo)
            time.sleep(0.5)
            
            browser.quit()  
             
     
```

```python
df = pd.DataFrame(list(zip(PhNo,Title)),columns=['Phone No','Title'])  
df.to_csv('IndiaCom.csv', index=False, encoding='utf-8') 
```
